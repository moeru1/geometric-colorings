```python
#criterio D(G)

def ady_vertices(e1, e2):
    (a, b) = e1
    (c, d) = e2
    #print(e1, e2)
    return a == c or a == d or b == c or b == d

def ady_D(s1, s2, e1, e2):
    return not (intersect(s1, s2) or ady_vertices(e1, e2))
```

```python
#Test de interseccion

def orientation(p, q, r):
    (px, py) = p
    (qx, qy) = q
    (rx, ry) = r
    a = qx-px
    b = rx-px
    c = qy-py
    d = ry-py
    return a*d-b*c


def ccw(a,b,c):
    (ax, ay) = a
    (bx, by) = b
    (cx, cy) = c
    return (cy-ay) * (bx-ax) > (by-ay) * (cx-ax)


# Return true if line segments AB and CD intersect
def intersect(s1, s2):
    (a, b) = s1
    (c, d) = s2
    inter  = ccw(a,c,d) != ccw(b,c,d) and ccw(a,b,c) != ccw(a,b,d)
    return inter
```

```python
#criterio Epp(G)

def ady_Epp(s1, s2, e1, e2):
    #return (not ady_vertices(e1, e2)) and intersect(s1, s2)
    return intersect(s1, s2)

def ady_Epp2(s1, s2, e1, e2):
    return intersect(s1,s2) and (not ady_vertices(e1, e2))
```

```python
#criterio W(G)

def ady_W(s1, s2, e1, e2):
    return not (ady_vertices(e1, e2)) or intersect(s1, s2) 
```

```python
from collections import deque
 
def adyacency_pairs(G):
    def print_pairs_dfs(v):
        stack = deque()
        stack.append(v)
        while stack:
            v = stack.pop()
            visited[v] = True
            for w in G.neighbor_iterator(v):
                if not visited[w]:
                    pairs.append((v,w))
                    stack.append(w)

    pairs = deque()
    #vertices 0 ... (n-1)
    n = len(G.vertices())
    #stack = deque()
    visited = [False] * n
    for v in G.vertices():
        if (not visited[v]):
            print_pairs_dfs(v)
            
    return pairs    
```

```python
def is_harmonious(G, coloring):
    S = set()
    pairs = adyacency_pairs(G)
    for (v,w) in pairs:
        c_v = coloring[v]
        c_w = coloring[w]
        if c_w != c_v and (not (frozenset((c_v, c_w)) in S)):
            S.add(frozenset((c_v, c_w)))
        else:
            return False
    return True
```

```python
def read_colorings(fname):
    with open(fname, "r") as f:
        i = 0
        h = []
        h_coloring = []
        for line in f:
            if (i % 5 == 0):
                name = line
            if (i % 5 == 1):
                h_line = line
                #print(h_line, end="")
                (_,h_val) = h_line.split("=")
                h_val = int(h_val)
                h.append(h_val)
            if (i % 5 == 2):
                coloring = list(map(int, line.split()))
                h_coloring.append(coloring)
            if (i % 5 == 3):
                diameter = line
            if (i % 5 == 4):
                None
            i += 1
        return h, h_coloring
```

```python
from collections import defaultdict

def draw_points(p_set, coloring):
    n = len(p_set)
    line_arr = []
    c = 0
    for i in range(n):
        for j in range(i+1, n):
            line_arr.append(line([p_set[i], p_set[j]], rgbcolor=color_list[coloring[c]]))
            c += 1
    
    draw = line_arr[0]
    for i in range(1, len(line_arr)):
        draw += line_arr[i]

    for i in range(n):
        draw += text(str(i), p_set[i])
    return draw


def draw_coloring(G, coloring):
    n = len(G.vertices())
    E = [(a,b) for a in range(n) for b in range(a+1, n)]
    #G = A(p_set, E)
    color_map = defaultdict(list)
    i = 0
    for p in G.vertices():
        color = color_list[coloring[i]]
        color_map[color].append(p)
        i += 1
    P = G.graphplot(vertex_colors = color_map)
    return P
```
